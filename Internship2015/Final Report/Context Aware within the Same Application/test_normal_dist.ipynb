```python
#visualization of the two dimention points
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
def  twoD_visual(z_label,n_clusters_,cluster_centers_indices):
     plt.figure(figsize=(10,10))
     plt.xlim(-5,25)
     plt.ylim(-5,35)
     plt.xlabel("x",fontsize=20)
     plt.ylabel("y",fontsize=20)
     for k,col in zip(range(n_clusters_),colors):
         for i in range(len(z_label)):
             if z_label[i][2]==k:
                  #plt.plot(z_label[i][0],z_label[i][1], 'o', markerfacecolor=col,
                      # markeredgecolor='black', markersize=5)
                 plt.text(z_label[i][0],z_label[i][1],k+1,color=col)
         plt.plot(z_label[cluster_centers_indices[k]][0],z_label[cluster_centers_indices[k]][1],
                          'o',markerfacecolor="red",markeredgecolor='black',markersize=10)
         print("%d Exemplar:(%f,%f)" % (k+1,z_label[cluster_centers_indices[k]][0],z_label[cluster_centers_indices[k]][1]))
     plt.title("Affinity Propagation Visualization",fontsize=15)
     for m in range(n_clusters_):
        plt.text(z_label[cluster_centers_indices[m]][0],z_label[cluster_centers_indices[m]][1],m+1,fontsize=20)
```

# First Batch with means(0,0),(10,10),(20,20)

```python
#generate the first batch of normal distribution with three means (0,0)(10,10)(20,20)
mean1 = [0,0]
cov1 = [[1,0],[0,1]]
mean2=[10,10]
cov2=[[1,0],[0,1]]
mean3=[20,20]
cov3=[[1,0],[0,1]]
x1,y1=np.random.multivariate_normal(mean1,cov1,100).T
x2,y2=np.random.multivariate_normal(mean2,cov2,100).T
x3,y3=np.random.multivariate_normal(mean3,cov3,100).T
x=np.concatenate((np.concatenate((x1,x2)),x3))
y=np.concatenate((np.concatenate((y1,y2)),y3))
x=np.array(x)
y=np.array(y)
z1=np.array(zip(x,y))
ap=AffinityPropagation().fit(z1) 
a1,b1,c1,d1=init_ap(ap,z1)
labels1=ap.labels_
z_labels1=zip(z1[:,0],z1[:,1],labels1)
cluster_indices1=ap.cluster_centers_indices_
cluster_centers1=ap.cluster_centers_
n1_cluster=len(cluster_centers1)
twoD_visual(z_labels1,n1_cluster,cluster_indices1)
plt.title(" First Batch AP Visual(means:(0,0),(10,10),(20,20))",fontsize=15)
```
